##### Kmeans CLustering needs numerical columns. 
##### We ned to engineer and generate more numerical columns like: 
##### Total spending: Quantity * Price
##### Total Orders by customers
##### Average spent (by each customer) =  Total_spending / Total Orders
##### Total Quantity (of each customer)
##### Frequency of Purchase = How many times each customer came

```python
data3['Customer ID'] = data3['Customer ID'].astype(str)
data3['InvoiceData'] = pd.to_datetime(data3['InvoiceDate'])
```

```python
data3 = data3[data3['Quantity'] >0 ]
```

```python
data3.shape
```

```python
data = data3.groupby('Customer ID').agg(
    Total_Spending=('Price', lambda x: (x * data3.loc[x.index, 'Quantity']).sum()),
    Total_Orders=('Invoice', 'nunique'),
    Total_Quantity=('Quantity', 'sum'),
    Avg_Spend=('Price', 'mean'),
    Distinct_Items=('StockCode', 'nunique'),
    First_Purchase=('InvoiceDate', 'min'),
    Last_Purchase=('InvoiceDate', 'max')
).reset_index()
```

```python
data.head(20)
```

```python
data.shape
```

```python
latest_purchase = data3['InvoiceDate'].max()



data['Last_Purchase'] = pd.to_datetime(data['Last_Purchase'], errors='coerce')
latest_purchase = pd.to_datetime(latest_purchase, errors='coerce')

data['Recency'] = (latest_purchase - data['Last_Purchase']).dt.days
```

```python
print(data[['Total_Spending', 'Total_Quantity', 'Distinct_Items', 'Recency']].describe())
```

```python
# Replace negative values with 0
data['Total_Spending'] = data['Total_Spending'].clip(lower=0)
data['Total_Quantity'] = data['Total_Quantity'].clip(lower=0)
data['Distinct_Items'] = data['Distinct_Items'].clip(lower=0)
data['Recency'] = data['Recency'].clip(lower=0)
data.fillna(0, inplace=True) 
```

```python
data['Total_Spending'] = np.log1p(data['Total_Spending'])
data['Total_Quantity'] = np.log1p(data['Total_Quantity'])
data['Distinct_Items'] = np.log1p(data['Distinct_Items'])
data['Recency'] = np.log1p(data['Recency'])
```

```python
x = data[['Total_Spending', 'Total_Orders', 'Total_Quantity', 
                          'Avg_Spend', 'Distinct_Items', 'Recency']]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)
```

```python
wcss = []

for i in range(1,10):
    kmeans = KMeans(i)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
    
wcss
```

```python
plt.plot(range(1,10), wcss)
plt.xlabel("number of clusters")
plt.ylabel('wcss')
```

```python
kmeans = KMeans(n_clusters=7, random_state=42)
data['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
identified_clusters = kmeans.fit_predict(X_scaled)
identified_clusters
```

```python
data_with_clusters = data.copy()
data_with_clusters['Cluster'] = identified_clusters
data_with_clusters
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

# Reduce to 2D for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Add PCA components to the DataFrame
data_with_clusters['PCA1'] = X_pca[:, 0]
data_with_clusters['PCA2'] = X_pca[:, 1]

# Scatter plot of clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', palette='viridis', data=data, alpha=0.6)
plt.title('Customer Segments (K-Means Clustering)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title="Cluster")
plt.show()
```

```python
# Get PCA component contributions
pca_components = pd.DataFrame(pca.components_, columns=x.columns, index=['PCA1', 'PCA2'])
print(pca_components)
```

```python
from sklearn.metrics import silhouette_score

sil_score = silhouette_score(X_scaled, kmeans.labels_)
print(f'Silhouette Score: {sil_score}')
```

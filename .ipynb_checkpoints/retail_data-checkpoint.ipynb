##### Kmeans CLustering needs numerical columns. 
##### We ned to engineer and generate more numerical columns like: 
##### Total spending: Quantity * Price
##### Total Orders by customers
##### Average spent (by each customer) =  Total_spending / Total Orders
##### Total Quantity (of each customer)
##### Frequency of Purchase = How many times each customer came

```python
data3['Customer ID'] = data3['Customer ID'].astype(str)
data3['InvoiceData'] = pd.to_datetime(data3['InvoiceDate'])
```

```python
data3 = data3[data3['Quantity'] >0 ]
```

```python
data3.shape
```

```python
data3.head(5)
```

```python
data = data3.groupby(['Customer ID', 'Country']).agg(
    Total_Spending=('Price', lambda x: (x * data3.loc[x.index, 'Quantity']).sum()),
    Total_Orders=('Invoice', 'nunique'),
    Total_Quantity=('Quantity', 'sum'),
    Distinct_Items=('StockCode', 'nunique'),
    First_Purchase=('InvoiceDate', 'min'),
    Last_Purchase=('InvoiceDate', 'max')
).reset_index()

data['Country'] = data['Country'].astype('category').cat.codes

country_mapping = dict(enumerate(data3['Country'].astype('category').cat.categories))
print(country_mapping)
```

```python
data.head(20)
```

```python
data.shape
```

```python
latest_purchase = data3['InvoiceDate'].max()



data['Last_Purchase'] = pd.to_datetime(data['Last_Purchase'], errors='coerce')
data['First_Purchase'] = pd.to_datetime(data['First_Purchase'], errors='coerce')
latest_purchase = pd.to_datetime(latest_purchase, errors='coerce')

data['Recency'] = (latest_purchase - data['Last_Purchase']).dt.days
```

```python
# data.drop(columns=['First_Purchase', 'Last_Purchase'], inplace=True)
data['Avg_Order_Value'] = data['Total_Spending'] / data['Total_Orders']
data['Avg_Items_Per_Order'] = data['Total_Quantity'] / data['Total_Orders']
# data['Frequency'] = data['Total_Orders'] / ((data['Last_Purchase'] - data['First_Purchase']).dt.days / 30)
```

```python
# from scipy.stats.mstats import winsorize
# data_wins = data_log.copy()
# data_wins['Total_Spending_Winsorized'] = winsorize(data_wins['Total_Spending'], limits=[0, 0.5])  # Cap top 1%
# data_wins['Total_Quantity_Winsorized'] = winsorize(data_wins['Total_Quantity'], limits=[0, 0.5]) 
# data_wins['Avg_Items_Per_Order_Winsorized'] = winsorize(data_wins['Avg_Items_Per_Order'], limits=[0, 0.5]) 
# data_wins['Avg_Order_Value  _Winsorized'] = winsorize(data_wins['Avg_Order_Value'], limits=[0, 0.1]) 
# data_wins['Total_Orders_Winsorized'] = winsorize(data_wins['Total_Orders_Order'], limits=[0, 0.1]) 


cols_to_winsorize = ['Total_Spending', 'Total_Orders', 'Total_Quantity', 
                      'Distinct_Items', 'Avg_Items_Per_Order', 'Avg_Order_Value']

data_log['Recency'] = data_log['Recency'].replace(-np.inf, np.nan)

#Ensure all columns are numeric
for col in cols_to_winsorize:
    data_log[col] = pd.to_numeric(data_log[col], errors='coerce')

# Winsorize: Cap values at the 80th percentile
for col in cols_to_winsorize:  
    q3 = data_log[col].quantile(0.80)
    data_log[col] = np.where(data_log[col] > q3, q3, data_log[col])
```

```python
x = data_log.copy()

features = ['Total_Spending', 'Total_Orders', 'Total_Quantity', 
                          'Avg_Order_Value', 'Avg_Items_Per_Order', 'Distinct_Items', 'Recency']
```

```python
# from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

    
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(x[features])

pca = PCA(n_components=2)
x_pca = pca.fit_transform(X_scaled)

X_scaled_ = pd.DataFrame(x_pca, columns=['PCA1', 'PCA2'])


```

```python
wcss = []

for i in range(1,10):
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10)
    kmeans.fit(X_scaled_)
    wcss.append(kmeans.inertia_)
    
wcss
```

```python
plt.plot(range(1,10), wcss)
plt.xlabel("number of clusters")
plt.ylabel('wcss')
```

```python
kmeans = KMeans(n_clusters=6, random_state=42, n_init=10)
identified_clusters = kmeans.fit_predict(X_scaled_)
x['Cluster'] = identified_clusters
X_scaled_['Cluster'] = identified_clusters
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='PCA1', y='PCA2', hue='Cluster', palette='viridis', 
    data=X_scaled_, alpha=0.6
)
plt.title('Customer Segments (K-Means Clustering)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title="Cluster")
plt.show()

```

```python
# Get PCA component contributions

#X_scaled_df = pd.DataFrame(X_scaled, columns=x.columns)

pca_components = pd.DataFrame(pca.components_, columns=features, index=['PCA1', 'PCA2'])
print(pca_components)
```

```python
from sklearn.metrics import silhouette_score

sil_score = silhouette_score(X_scaled_, identified_clusters)
print(f'Silhouette Score: {sil_score}')
```

```python
cluster_summary = x.groupby("Cluster")[features].mean()
print(cluster_summary)
```

```python

x['Country'] = x['Country'].map(country_mapping).fillna('Unknown')


country_cluster_distribution = x.groupby(['Country', 'Cluster']).size().unstack()


print(country_cluster_distribution.to_string())  
```

```python
import matplotlib.pyplot as plt


top_n = 5  
country_totals = country_cluster_distribution.sum(axis=1)
top_countries = country_totals.nlargest(top_n).index


filtered_data = country_cluster_distribution.loc[top_countries]
filtered_data.loc["Other"] = country_cluster_distribution.loc[~country_cluster_distribution.index.isin(top_countries)].sum()

# Plot stacked bar chart
filtered_data.plot(kind="bar", figsize=(15, 7), stacked=True, colormap="viridis")

plt.xlabel("Country")
plt.ylabel("Number of Customers")
plt.title("Customer Distribution per Country by Cluster")
plt.legend(title="Cluster")
plt.xticks(rotation=90)
plt.show()
```

##### PCA1 is positively correlated with Total Spending, Total Orders, Quantity, and Distinct Items Value
##### PCA2 appear to represent Recency

#####  Cluster 0: "Frequent Big Spenders"
##### High Spending, High Orders, High Variety
##### Low Recency (very recent customers)
##### Business Tip: Offer exclusive deals, premium memberships, or discounts to retain them.

#####  Cluster 1: "Occasional Buyers"
##### Average Spendings, Low Orders, Low Variety(Low Distinct Items)
##### High Recency (older customers)
##### Business Tip: Send reminders, discounts, or personalized recommendations to re-engage them.

#####  Cluster 2: "Balanced or Regular Buyers"
##### Average Spendings, Average Orders, Average Variety(Average Distinct Items)
##### Low Recency (Recentcustomers)
##### Business Tip:  Offer a subscription model or loyalty rewards to increase their frequency.

#####  Cluster 3: "Loyal but Rare Buyers"
##### Average Spendings, Low Orders, Average Variety(Average Distinct Items)
##### High Recency (Old customers)
##### spend well when they buy
##### Business Tip: Offer periodic check-in emails or reminders for seasonal discounts.

#####  Cluster 4: "Inactive Customers"
##### Low Spending, Low Orders, Very Low Variety
##### Highest Recency (Oldest customers)
##### Business Tip: Consider re-engagement campaigns, free shipping offers, or surveys to understand why they left.

#####  Cluster 5: "High-Value Customers"
##### High Spending, Average Orders, High Variety
##### High Recency (not very recent customers)
##### Spend a lot, buy different items, but not as often as Cluster 0.
##### Business Tip: Offer VIP perks, early access to new products, or personalized thank-you emails.
